**1. Formal Definition and Notational Setup**

Let $\mathbf{Z} \in \mathbb{R}^{n+m}$ represent a partitioned random vector decomposed into two sub-vectors $\mathbf{X} \in \mathbb{R}^n$ and $\mathbf{Y} \in \mathbb{R}^m$:$$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix}$$The vector $\mathbf{Z}$ follows a Joint Multivariate Normal Distribution if and only if every linear combination of its elements yields a univariate normal distribution. Assuming that the distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$, its covariance matrix must be symmetric and strictly positive-definite (non-singular).The joint probability density function (PDF) $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ is expressed as:$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{(2\pi)^{\frac{n+m}{2}} |\boldsymbol{\Sigma}|^{\frac{1}{2}}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right)$$where the parameters are structured hierarchically:Expectation Vector ($\boldsymbol{\mu}$):$$\boldsymbol{\mu} = \mathbb{E}[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix}$$Partitioned Covariance Matrix ($\boldsymbol{\Sigma}$):$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$$\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ sizing $n \times n$$\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ sizing $m \times m$$\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ sizing $n \times m$

**2. Derivation of Marginal Distributions via Block Matrix Factorization**

To break down the joint distribution into its constituent marginal profiles, we integrate out the unwanted variables. For instance, to acquire $f_{\mathbf{X}}(\mathbf{x})$, we compute:$$f_{\mathbf{X}}(\mathbf{x}) = \int_{\mathbb{R}^m} f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) \, d\mathbf{y}$$To make this analytically tractable, we decompose the joint quadratic exponent using the Schur Complement. The inverse of our partitioned covariance matrix can be factored as:$$\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \mathbf{I}_n & \mathbf{0} \\ -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} & \mathbf{I}_m \end{bmatrix}^T \begin{bmatrix} \mathbf{S}_{XX}^{-1} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \mathbf{I}_n & \mathbf{0} \\ -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} & \mathbf{I}_m \end{bmatrix}$$where $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$ is the Schur complement of $\boldsymbol{\Sigma}_{YY}$.

**2.1 Isolating the Marginal Density of $\mathbf{X}$**

By shifting the coordinate system to zero-centered variables $\widetilde{\mathbf{x}} = \mathbf{x} - \boldsymbol{\mu}_X$ and $\widetilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_Y$, the quadratic exponent $Q$ expands elegantly into two independent components:$$Q = \widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}} + (\widetilde{\mathbf{y}} - \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}\widetilde{\mathbf{x}})^T \mathbf{S}_{YY}^{-1} (\widetilde{\mathbf{y}} - \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}\widetilde{\mathbf{x}})$$Factoring out the term that depends exclusively on $\mathbf{x}$ from the integral yields:$$f_{\mathbf{X}}(\mathbf{x}) = \frac{1}{(2\pi)^{\frac{n+m}{2}}|\boldsymbol{\Sigma}|^{\frac{1}{2}}} \exp\left(-\frac{1}{2}\widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1}\widetilde{\mathbf{x}}\right) \int_{\mathbb{R}^m} \exp\left(-\frac{1}{2}(\widetilde{\mathbf{y}}-\mathbf{m})^T \mathbf{S}_{YY}^{-1}(\widetilde{\mathbf{y}}-\mathbf{m})\right) d\mathbf{y}$$Because the integrand is the unnormalized kernel of a Gaussian distribution vector centered around $\mathbf{m} = \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}\widetilde{\mathbf{x}}$, the integral evaluates to its normalization constant: $\sqrt{(2\pi)^m |\mathbf{S}_{YY}|}$.Applying the matrix determinant lemma $| \boldsymbol{\Sigma} | = | \boldsymbol{\Sigma}_{XX} | \cdot | \mathbf{S}_{YY} |$, the expression collapses cleanly into:$$f_{\mathbf{X}}(\mathbf{x}) = \frac{1}{(2\pi)^{\frac{n}{2}} |\boldsymbol{\Sigma}_{XX}|^{\frac{1}{2}}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}_X)^T \boldsymbol{\Sigma}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}_X) \right)$$This confirms that the marginal distribution preserves normality: $\mathbf{X} \sim \mathcal{N}(\boldsymbol{\mu}_X, \boldsymbol{\Sigma}_{XX})$. By symmetry, $\mathbf{Y} \sim \mathcal{N}(\boldsymbol{\mu}_Y, \boldsymbol{\Sigma}_{YY})$.

**3. Derivation of the Conditional Density**

Using the foundational definition of continuous conditional probability distributions, we write:$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$Substituting the joint and marginal density functions requires subtracting the exponent of the marginal distribution from the joint quadratic form.

**3.1 Exponent Algebraic Subtraction**

The joint quadratic expression can be grouped relative to $\mathbf{Y}$:$$Q_{\text{joint}} = (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) + (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$where $\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$.Subtracting the marginal exponent $Q_{\text{marginal}} = (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y)$ leaves only the second term:$$Q_{\text{conditional}} = (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$

**3.2 Evaluating the Scaling Constant**

Dividing the normalization constants yields:$$\frac{(2\pi)^{\frac{m}{2}} |\boldsymbol{\Sigma}_{YY}|^{\frac{1}{2}}}{(2\pi)^{\frac{n+m}{2}} |\boldsymbol{\Sigma}|^{\frac{1}{2}}} = \frac{1}{(2\pi)^{\frac{n}{2}} \left( \frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|} \right)^{\frac{1}{2}}}$$Invoking the identity $\frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|} = |\mathbf{S}_{XX}|$, we achieve the final conditional PDF profile:$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{(2\pi)^{\frac{n}{2}} |\boldsymbol{\Sigma}_{X|Y}|^{\frac{1}{2}}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}_{X|Y})^T \boldsymbol{\Sigma}_{X|Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}_{X|Y}) \right)$$Hence, the conditional distribution is explicitly written as:$$\mathbf{X} \mid \mathbf{Y} = \mathbf{y} \sim \mathcal{N}(\boldsymbol{\mu}_{X|Y}, \boldsymbol{\Sigma}_{X|Y})$$Parameter SummaryConditional Expectation: $\boldsymbol{\mu}_{X|Y} = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$Conditional Covariance Matrix: $\boldsymbol{\Sigma}_{X|Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$

**4. Analysis of Conditional Expectation and Variance Decomposition**

**4.1 Conditional Expectation Function**

By the Disintegration Theorem, the conditional expectation operator acts as the first moment of the conditional density function:$$\mathbb{E}[\mathbf{X} \mid \mathbf{Y}=\mathbf{y}] = \int_{\mathbb{R}^n} \mathbf{x} f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) \, d\mathbf{x} = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$$Treating the observation as a random vector maps this to a linear transformation of $\mathbf{Y}$:$$\mathbb{E}[\mathbf{X} \mid \mathbf{Y}] = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{Y} - \boldsymbol{\mu}_Y)$$

**4.2 The Law of Total Variance**

The Law of Total Variance states that the global dispersion of a random variable can be broken down into two components:$$\text{Var}(\mathbf{X}) = \mathbb{E}[\text{Var}(\mathbf{X} \mid \mathbf{Y})] + \text{Var}(\mathbb{E}[\mathbf{X} \mid \mathbf{Y}])$$Unexplained (Residual) Variance:$$\mathbb{E}[\text{Var}(\mathbf{X} \mid \mathbf{Y})] = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$$This captures the residual uncertainty left in $\mathbf{X}$ after measuring $\mathbf{Y}$. Crucially, in Gaussian systems, this value is entirely independent of the observed value $\mathbf{y}$.Explained Variance:$$\text{Var}(\mathbb{E}[\mathbf{X} \mid \mathbf{Y}]) = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$$This quantifies the variance reduction or information gained due to the linear cross-dependencies between $\mathbf{X}$ and $\mathbf{Y}$.

**5. Information-Theoretic Metrics**

**5.1 Joint Differential Entropy**

The differential entropy $h(\mathbf{Z})$ evaluates the expected log-uncertainty of the joint system:$$h(\mathbf{Z}) = -\mathbb{E}[\ln f_{\mathbf{X},\mathbf{Y}}(\mathbf{z})] = \frac{1}{2}\ln\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right) + \frac{1}{2}\mathbb{E}\left[(\mathbf{Z}-\boldsymbol{\mu})^T\boldsymbol{\Sigma}^{-1}(\mathbf{Z}-\boldsymbol{\mu})\right]$$Since the expectation of the quadratic form corresponds to $\text{Tr}(\boldsymbol{\Sigma}^{-1}\boldsymbol{\Sigma}) = \text{Tr}(\mathbf{I}_{n+m}) = n+m$, we can rewrite this as:$$h(\mathbf{Z}) = \frac{1}{2} \ln \left( (2\pi e)^{n+m} |\boldsymbol{\Sigma}| \right)$$

**5.2 Conditional Differential Entropy (Corrected)**

The conditional entropy $h(\mathbf{X} \mid \mathbf{Y})$ reflects the remaining uncertainty of $\mathbf{X}$ given knowledge of $\mathbf{Y}$. It relies on the true conditional covariance matrix $\boldsymbol{\Sigma}_{X|Y}$:$$h(\mathbf{X} \mid \mathbf{Y}) = \frac{1}{2} \ln \left( (2\pi e)^n |\boldsymbol{\Sigma}_{X|Y}| \right) = \frac{1}{2} \ln \left( (2\pi e)^n |\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}| \right)$$

**5.3 Mutual Information**

Mutual Information $I(\mathbf{X}; \mathbf{Y})$ measures the shared information content between the two variable blocks:$$I(\mathbf{X}; \mathbf{Y}) = h(\mathbf{X}) - h(\mathbf{X} \mid \mathbf{Y})$$Substituting our marginal and conditional entropy expressions causes the scaling factor $(2\pi e)^n$ to cancel out:$$I(\mathbf{X}; \mathbf{Y}) = \frac{1}{2} \ln \left( \frac{|\boldsymbol{\Sigma}_{XX}|}{|\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}|} \right)$$

# Numerical Demostration

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Parameter Settings
mean_vector = np.array([0, 0])
cov_matrix = np.array([[1.0, 0.75],
                        [0.75, 1.0]])  # Highly correlated Gaussian system

# Generate Synthetic Samples
rng = np.random.default_rng(seed=42)
num_samples = 6000
data_samples = rng.multivariate_normal(mean_vector, cov_matrix, size=num_samples)
x_data, y_data = data_samples[:, 0], data_samples[:, 1]

# 2. Compute Conditional Parameters at a Fixed Value of Y
fixed_y = 1.3
tolerance = 0.08  # Width of empirical slice

# Empirical filtering
conditional_mask = np.abs(y_data - fixed_y) < tolerance
x_conditional_empirical = x_data[conditional_mask]

# Theoretical conditional calculations
mu_x, mu_y = mean_vector[0], mean_vector[1]
sigma_xx, sigma_yy = cov_matrix[0, 0], cov_matrix[1, 1]
sigma_xy = cov_matrix[0, 1]

theoretical_mu_cond = mu_x + (sigma_xy / sigma_yy) * (fixed_y - mu_y)
theoretical_var_cond = sigma_xx - (sigma_xy**2 / sigma_yy)

# 3. Create the Visualization Layout
fig, axs = plt.subplots(2, 2, figsize=(11, 10))

# --- Top-Left: Joint Distribution Density Map ---
hb = axs[0, 0].hexbin(x_data, y_data, gridsize=35, cmap='plasma', mincnt=1)
axs[0, 0].axhline(y=fixed_y, color='cyan', linestyle='--', linewidth=2,
                   label=f'Conditioning Slice (Y={fixed_y})')
axs[0, 0].set_title("Joint Density Profile: $P(X, Y)$", fontsize=12)
axs[0, 0].set_xlabel("Variable X")
axs[0, 0].set_ylabel("Variable Y")
axs[0, 0].legend(loc='upper left')
fig.colorbar(hb, ax=axs[0, 0], label='Sample Counts')

# --- Top-Right: Marginal Distribution of Y ---
axs[0, 1].hist(y_data, bins=40, color='royalblue', alpha=0.65, density=True)
axs[0, 1].set_title("Marginal Distribution: $P(Y)$", fontsize=12)
axs[0, 1].set_xlabel("Y")
axs[0, 1].set_ylabel("Density")

# --- Bottom-Left: Marginal Distribution of X ---
axs[1, 0].hist(x_data, bins=40, color='forestgreen', alpha=0.65, density=True)
axs[1, 0].set_title("Marginal Distribution: $P(X)$", fontsize=12)
axs[1, 0].set_xlabel("X")
axs[1, 0].set_ylabel("Density")

# --- Bottom-Right: Conditional Distribution with Theoretical Overlay ---
axs[1, 1].hist(x_conditional_empirical, bins=25, color='crimson', alpha=0.6,
               density=True, label='Empirical Data Slice')

# Plot continuous theoretical curve
x_grid = np.linspace(theoretical_mu_cond - 3.5 * np.sqrt(theoretical_var_cond),
                     theoretical_mu_cond + 3.5 * np.sqrt(theoretical_var_cond), 200)
pdf_curve = (1.0 / np.sqrt(2 * np.pi * theoretical_var_cond)) * \
            np.exp(-(x_grid - theoretical_mu_cond)**2 / (2 * theoretical_var_cond))
axs[1, 1].plot(x_grid, pdf_curve, color='black', linestyle='-', linewidth=2.5,
               label='Theoretical Normal PDF')

axs[1, 1].set_title(f"Conditional Distribution: $P(X \mid Y={fixed_y})$", fontsize=12)
axs[1, 1].set_xlabel("X")
axs[1, 1].set_ylabel("Density")
axs[1, 1].legend(loc='upper right')

plt.tight_layout()
plt.savefig('multivariate_gaussian_analysis.png', dpi=300)
print("Plot successfully saved to disk.")
```

**Q3.1)Joint Distribution of $(X, Y)$**

 We are given the following standard normal components for the system:

 Hidden Temperature: $X \sim \mathscr{N}(0, 1)$Mechanical Load (Noise): $Q \sim \mathscr{N}(0, 4)$Independence: $X \perp Q$, which mathematically implies that their cross-covariance is zero: $\text{Cov}(X, Q) = 0$.The observed power sensor reading is defined by the linear combination:$$Y = X + Q$$

** Step A: Show that $Y \sim \mathscr{N}(0, 5)$**

 Because $X$ and $Q$ are independent normal random variables, any linear combination of them remains normally distributed.Expected Value (Mean): Using the linearity of the expectation operator:$$\mathbb{E}[Y] = \mathbb{E}[X + Q] = \mathbb{E}[X] + \mathbb{E}[Q] = 0 + 0 = 0$$Variance: Using the properties of variance for the sum of two random variables:$$\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) + 2\text{Cov}(X, Q)$$Since $X \perp Q$, the covariance term vanishes, yielding:$$\text{Var}(Y) = 1 + 4 + 0 = 5$$

** Step B: Find the Covariance $\text{Cov}(X, Y) = \boldsymbol{\Sigma}_{XY}$**

 Using the definition of covariance and substituting $Y = X + Q$:$$\text{Cov}(X, Y) = \mathbb{E}[XY] - \mathbb{E}[X]\mathbb{E}[Y]$$Since $\mathbb{E}[X] = 0$, this simplifies directly to the inner product expectation:$$\text{Cov}(X, Y) = \mathbb{E}[X(X + Q)] = \mathbb{E}[X^2 + XQ] = \mathbb{E}[X^2] + \mathbb{E}[XQ]$$By definition, $\mathbb{E}[X^2] = \text{Var}(X) + (\mathbb{E}[X])^2 = 1 + 0 = 1$.By independence, $\mathbb{E}[XQ] = \mathbb{E}[X]\mathbb{E}[Q] = 0 \cdot 0 = 0$.Therefore, the cross-covariance block is:$$\boldsymbol{\Sigma}_{XY} = \text{Cov}(X, Y) = 1$$Joint Matrix FormThe full joint distribution profile of the random vector can be stacked into a bivariate normal system:$$\begin{bmatrix} X \\ Y\end{bmatrix} \sim \mathscr{N} \left( \begin{bmatrix} 0 \\ 0 \end{bmatrix}, \begin{bmatrix} 1 & 1 \\ 1 & 5 \end{bmatrix} \right)$$

** Q3.2) Information Geometry and Measurability**

 Measure-Theoretic AnalysisA random variable $X$ is said to be $\sigma(Y)$-measurable if and only if the informational resolution generated by $Y$ can uniquely determine $X$. This requires the existence of a deterministic, Borel-measurable function $g: \mathbb{R} \to \mathbb{R}$ such that $X = g(Y)$.In this system, $X$ is not $\sigma(Y)$-measurable. The observation mapping $Y = X + Q$ compresses a two-dimensional continuous coordinate state space $(X, Q) \in \mathbb{R}^2$ down into a one-dimensional projection space $Y \in \mathbb{R}$. For any given realized sensor output $Y = y$, there exists an uncountably infinite set of pairs $(x, q)$ that satisfy the algebraic constraint $x + q = y$.Practical Engineering InterpretationIf the power meter displays a spike of $Y = 5$, the engineer faces an irreducible structural ambiguity. This specific value could be caused by:A severe thermal fault where the motor is melting ($X = 5$) while running completely free of mechanical load ($Q = 0$).A completely normal thermal state ($X = 0$) while trying to cut through an incredibly hard, dense node of raw material ($Q = 5$).Because the single sensor aggregates both physical phenomena into one metric, looking at $Y$ alone cannot isolate the underlying cause.

** Q3.3) Conditional Expectation (The Best Estimate)**

 For any bivariate normal distribution, the Minimum Mean Square Error (MMSE) estimator is proven to be the conditional expectation operator, which expands via the linear regression formula:$$\mathbb{E}[X \mid Y = y] = \boldsymbol{\mu}_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \boldsymbol{\mu}_Y)$$Substituting our derived structural parameters ($\boldsymbol{\mu}_X = 0$, $\boldsymbol{\mu}_Y = 0$, $\text{Cov}(X,Y) = 1$, and $\text{Var}(Y) = 5$) into this equation yields:$$\mathbb{E}[X \mid Y = y] = 0 + \frac{1}{5}(y - 0) = 0.2y$$Lifting this definition to a global random variable mapping across the sample space gives:$$\mathbb{E}[X \mid Y] = 0.2Y$$Signal-to-Total-Variance Ratio ExpressionWe can rewrite the scaling factor (the Kalman-like filter gain $K$) to expose its physical information composition:$$K = \frac{\text{Cov}(X,Y)}{\text{Var}(Y)} = \frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)} = \frac{\sigma_{\text{signal}}^2}{\sigma_{\text{signal}}^2 + \sigma_{\text{noise}}^2} = \frac{1}{1 + 4} = 0.2$$This ratio demonstrates that because the mechanical noise background ($\sigma^2 = 4$) is four times more volatile than the thermal signal ($\sigma^2 = 1$), we only "trust" $20\%$ of any raw fluctuation observed on the power meter as being truly indicative of a temperature shift.

** Q3.4) Verification of the Residual**

 Let $\widehat{X} = \mathbb{E}[X \mid Y] = 0.2Y$ denote our optimal estimator. The random estimation error vector (residual) is defined as:$$Z = X - \widehat{X} = X - 0.2Y$$Part A: Prove that $\mathbb{E}[Z] = 0$Using the linearity of the expectation operator:$$\mathbb{E}[Z] = \mathbb{E}[X - 0.2Y] = \mathbb{E}[X] - 0.2\mathbb{E}[Y]$$Substituting the baseline system expectations ($\mathbb{E}[X] = 0$ and $\mathbb{E}[Y] = 0$):$$\mathbb{E}[Z] = 0 - 0.2(0) = 0$$Part B: Verify Uncorrelatedness (The Orthogonality Principle)To prove that all available information has been extracted from our observation window, we check the cross-covariance between the residual error $Z$ and the sensor variable $Y$:$$\text{Cov}(Z, Y) = \mathbb{E}[ZY] - \mathbb{E}[Z]\mathbb{E}[Y]$$Since $\mathbb{E}[Z] = 0$, this reduces to the inner product expectation:$$\text{Cov}(Z, Y) = \mathbb{E}[(X - 0.2Y)Y] = \mathbb{E}[XY - 0.2Y^2] = \mathbb{E}[XY] - 0.2\mathbb{E}[Y^2]$$From our previous expansions, $\mathbb{E}[XY] = \text{Cov}(X, Y) = 1$.By definition, $\mathbb{E}[Y^2] = \text{Var}(Y) + (\mathbb{E}[Y])^2 = 5 + 0 = 5$.Substituting these values back into the equation:$$\text{Cov}(Z, Y) = 1 - 0.2(5) = 1 - 1 = 0$$Because their cross-covariance is exactly zero, the estimation error is perfectly orthogonal to the observation space in the Hilbert space of square-integrable random variables ($\text{Cov}(Z, Y) = 0$). This guarantees that no residual predictive power remains unexploited in $Y$.

** Q3.5) Information Theory: Differential Entropy**

 The differential entropy of a continuous random variable distributed according to a normal distribution $\mathscr{N}(\mu, \sigma^2)$ depends exclusively on its variance parameter:$$h(\mathscr{N}(\mu, \sigma^2)) = \frac{1}{2}\ln(2\pi e \sigma^2) \text{ nats}$$Step A: Compute the Marginal Entropy $h(X)$Using the core thermal signal variance $\sigma_X^2 = 1$:$$h(X) = \frac{1}{2}\ln(2\pi e \cdot 1) = \frac{1}{2}\ln(2\pi e) \approx 1.4189 \text{ nats}$$Step B: Compute the Conditional Differential Entropy $h(X \mid Y)$We first calculate the conditional variance $\sigma_{X|Y}^2$, which represents the remaining uncertainty spread after observing $Y$:$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1^2}{5} = 1 - 0.2 = 0.8$$Substituting this tightened variance profile into the continuous entropy metric:$$h(X \mid Y) = \frac{1}{2}\ln(2\pi e \sigma_{X|Y}^2) = \frac{1}{2}\ln(2\pi e \cdot 0.8) \approx 1.3073 \text{ nats}$$

** Q3.6) Mutual Information**

 The mutual information measures the absolute reduction in spatial uncertainty achieved by deploying our power meter channel:$$I(X; Y) = h(X) - h(X \mid Y)$$Substituting our logarithmic closed-form definitions from Q3.5:$$I(X; Y) = \frac{1}{2}\ln(2\pi e) - \frac{1}{2}\ln(2\pi e \cdot 0.8) = \frac{1}{2}\ln\left( \frac{2\pi e}{2\pi e \cdot 0.8} \right) = \frac{1}{2}\ln\left(\frac{1}{0.8}\right) = \frac{1}{2}\ln(1.25) \approx 0.1116 \text{ nats}$$Uncertainty Metrics AnalysisPercentage Uncertainty Removed: In continuous Gaussian structures, informational gain is tracked by the relative contraction of variance. The system variance drops from an initial $\sigma_X^2 = 1.0$ down to a conditional $\sigma_{X|Y}^2 = 0.8$. This represents a definitive $20\%$ reduction in uncertainty ($1 - 0.8 = 0.20$).Asymptotic Noise Scenario ($\sigma_Q^2 = 100$): If mechanical noise expands to $100$, our variance components alter the system:$$\text{Var}(Y) = 1 + 100 = 101 \implies \sigma_{X|Y}^2 = 1 - \frac{1^2}{101} = \frac{100}{101} \lim_{\sigma_Q^2 \to \infty} \sigma_{X|Y}^2 = 1.0$$As a result, the mutual information equations collapse:$$I(X;Y) = \frac{1}{2}\ln\left(\frac{1}{100/101}\right) = \frac{1}{2}\ln(1.01) \to 0 \text{ nats}$$The temperature signal becomes completely drowned out by the mechanical noise, rendering the sensor useless.

** Q3.7) Diagnostic Interpretation and Shrinkage**

 The manager’s proposal to assume $\widehat{X}_{\text{naive}} = Y$ represents a classic Naive Estimator that ignores the presence of background noise.Bias and Sub-optimality ProofIf we implement the naive approach, our actual error variance expands dramatically:$$\text{MSE}_{\text{naive}} = \mathbb{E}[(X - Y)^2] = \mathbb{E}[(X - (X + Q))^2] = \mathbb{E}[(-Q)^2] = \text{Var}(Q) = 4.0$$In comparison, our optimized conditional expectation yields an error variance equal to the conditional variance derived in Q3.5:$$\text{MSE}_{\text{optimal}} = \sigma_{X|Y}^2 = 0.8$$The naive approach results in 5 times more error ($4.0$ vs $0.8$) than the optimal model.

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# --- 1. Parameters for the Fraud Detection Problem ---
# X: Fraud Indicator (0 = Legitimate, 1 = Fraudulent)
# Y: Customer Risk Class (0 = Low-Risk, 1 = High-Risk)
# Joint Probabilities:
# P(X=0, Y=0) = 0.50, P(X=1, Y=0) = 0.10 => P(Y=0) = 0.60
# P(X=0, Y=1) = 0.20, P(X=1, Y=1) = 0.20 => P(Y=1) = 0.40
p_w1, p_w2, p_w3, p_w4 = 0.50, 0.10, 0.20, 0.20

# Conditional Expectations E[X|Y]
ey0 = 1 / 6  # E[X|Y=0] = 0.10 / 0.60
ey1 = 1 / 2  # E[X|Y=1] = 0.20 / 0.40
global_mean = 0.30  # E[X]

# --- 2. Generate Simulated Data for Visualization ---
np.random.seed(42)
n_samples = 1000
outcomes = np.random.choice([1, 2, 3, 4], size=n_samples, p=[p_w1, p_w2, p_w3, p_w4])

# Map outcomes to X and Y
x_samples = np.where((outcomes == 2) | (outcomes == 4), 1, 0)
y_samples = np.where((outcomes == 3) | (outcomes == 4), 1, 0)

# Compute optimal estimator and residuals
x_hat = np.where(y_samples == 0, ey0, ey1)
residual = x_samples - x_hat

# Add jitter to discrete values for a beautiful data cloud visualization
jitter_x = x_samples + np.random.uniform(-0.08, 0.08, n_samples)
jitter_y = y_samples + np.random.uniform(-0.08, 0.08, n_samples)
jitter_y_res = y_samples + np.random.uniform(-0.08, 0.08, n_samples)

# --- 3. Create Simulation Plot ---
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "Joint Distribution: X vs Y<br><sup>The Discrete Best Estimate (Conditional Expectation)</sup>",
        "Estimation Error (Residual) vs Y<br><sup>Visualizing the Orthogonality Principle</sup>",
    ),
)

# --- LEFT PLOT: Joint X, Y and the Estimator ---
# Scatter of actual realizations with jitter to see density
fig.add_trace(
    go.Scatter(
        x=jitter_x,
        y=jitter_y,
        mode="markers",
        marker=dict(color="rgba(31, 119, 180, 0.25)", size=4),
        name="Transactions (Jittered)",
    ),
    row=1,
    col=1,
)

# The Best Estimate Line E[X|Y]
y_line = np.array([0, 1])
x_line_optimal = np.array([ey0, ey1])
fig.add_trace(
    go.Scatter(
        x=x_line_optimal,
        y=y_line,
        mode="lines+markers",
        marker=dict(size=10),
        line=dict(color="Red", width=4),
        name="E[X|Y] (Optimal MMSE)",
    ),
    row=1,
    col=1,
)

# The Naive/Global Average Line X = 0.30
fig.add_trace(
    go.Scatter(
        x=[global_mean, global_mean],
        y=[-0.2, 1.2],
        mode="lines",
        line=dict(color="Black", width=2, dash="dash"),
        name="X = 0.30 (Global Mean)",
    ),
    row=1,
    col=1,
)

# --- RIGHT PLOT: Orthogonality (Residual vs Y) ---
fig.add_trace(
    go.Scatter(
        x=jitter_y_res,
        y=residual,
        mode="markers",
        marker=dict(color="Purple", opacity=0.4, size=4),
        name="Residual (X - E[X|Y])",
    ),
    row=1,
    col=2,
)

# Horizontal line at 0 to show zero mean error
fig.add_trace(
    go.Scatter(
        x=[-0.2, 1.2],
        y=[0, 0],
        mode="lines",
        line=dict(color="Black", width=2),
        name="Zero Mean Error",
    ),
    row=1,
    col=2,
)

# --- Layout and Styling ---
fig.update_layout(
    height=600,
    title_text="Credit Card Fraud Monitoring Simulation: Extracting Signal from Risk Class",
    template="plotly_white",
    hovermode="x",
)

fig.update_xaxes(title_text="Fraud Indicator (X)", range=[-0.2, 1.2], row=1, col=1)
fig.update_yaxes(
    title_text="Customer Risk Class (Y)", range=[-0.2, 1.2], row=1, col=1
)
fig.update_xaxes(
    title_text="Customer Risk Class (Y)", range=[-0.2, 1.2], row=1, col=2
)
fig.update_yaxes(title_text="Residual Error (Z)", range=[-1.0, 1.0], row=1, col=2)

fig.show()
```
